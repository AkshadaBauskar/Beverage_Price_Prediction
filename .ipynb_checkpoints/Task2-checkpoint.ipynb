#### Analyze the total quantity sold after promotion for the Sankranti campaign across different product categories. What percentage does each category contribute to the overall sales, and what insights can be drawn from these contributions?
 

```python
df_products_events = pd.merge(df_dim_products, df_fact_events, on ='product_code')
df_products_events.head(3)
```

```python
df_sankranti = df_products_events[df_products_events['campaign_id'] == 'CAMP_SAN_01']
df_sankranti.head(3)
```

```python
df_sankranti.groupby('category')['quantity_sold(after_promo)'].sum()
```

```python
total_sales = df_sankranti['quantity_sold(after_promo)'].sum()
total_sales
```

```python
per_sales = (df_sankranti.groupby('category')['quantity_sold(after_promo)'].sum()) * 100/ total_sales
per_sales
```

Based on the pie chart:

Percentage Contribution of Each Category:

Grocery & Staples: 70.5% - The largest contributor to the total sales after the Sankranti promotion.
Home Appliances: 14.1% - The second-highest contributor.
Home Care: 6.7%.
Personal Care: 4.9%.
Combo1: 3.7% - The smallest contributor.
Insights:

Grocery & Staples dominate the sales, indicating that this category is highly popular or heavily promoted during the campaign. It likely aligns well with customers' needs during the Sankranti festival.
Home Appliances contribute significantly, possibly due to festival-related purchases or attractive discounts.
Other categories like Home Care, Personal Care, and Combo1 have much smaller contributions, suggesting either lower demand or less aggressive promotion.

#### Examine the correlation between base price (after the promotion) and sales quantities (after the promotion). What insights can be drawn regarding the relationship between base price and sales quantities after the promotion?

```python
correlation = df_fact_events[['base_price(after_promo)', 'quantity_sold(after_promo)']].corr()
correlation
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Between Base Price and Sales Quantity (After Promotion)')
plt.show()
```

*Insights from the Heatmap:*
Correlation Value:

The correlation coefficient between base_price(after_promo) and quantity_sold(after_promo) is 0.27.
This indicates a weak positive correlation, suggesting that as the base price increases, the sales quantity slightly increases after the promotion.
Implications:

The weak positive correlation might be due to:
Higher base-priced products being perceived as premium, leading to better sales even after a small promotion.
Certain customers being less price-sensitive and willing to purchase even at higher prices.
The weak nature of the correlation suggests that price is not the sole determinant of sales quantities, and other factors (e.g., product quality, brand reputation, or the nature of the promotion) might have a stronger influence.

#### Analyze the incremental sold units percentage (ISU%) across various cities. Identify the city with the highest ISU% after the promotion and the city with the smallest change. What trends can be observed about the effectiveness of promotions in driving sales across these cities?

```python
quantity_bef = df_stores_events.groupby('city')['quantity_sold(before_promo)'].sum()
quantity_aft = df_stores_events.groupby('city')['quantity_sold(after_promo)'].sum()

quantity_diff = quantity_aft - quantity_bef
quantity_diff

ISU = quantity_diff * 100 / quantity_bef
ISU
```

```python
ISU_sorted = ISU.sort_values(ascending= True)
ISU_sorted
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x = ISU_sorted.index, y = ISU_sorted.values, marker='o', color='b')
plt.title('Incremental Sold Units Percentage (ISU%) Across Cities', fontsize=16)
plt.xlabel('City', fontsize=12)
plt.ylabel('ISU% (Incremental Sold Units Percentage)', fontsize=12)

# Rotate x-axis labels for readability
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()
plt.show()
```

Observations from the ISU% Chart:
City with the Highest ISU%:

Madurai has the highest ISU%, indicating that the promotion was most effective in driving incremental sales in this city.
This suggests strong promotional responsiveness, likely due to high demand or effective alignment with customer preferences.
City with the Smallest ISU%:

Visakhapatnam has the lowest ISU%, suggesting minimal impact of the promotion on sales.
This could be due to market saturation, ineffective promotions, or a lower demand for the promoted products.
Overall Trends:

There is a clear upward trend in ISU% from Visakhapatnam to Madurai, indicating varying promotional effectiveness across cities.
The cities with higher ISU% (e.g., Madurai, Bengaluru, and Coimbatore) likely have a more price-sensitive or promotion-responsive customer base.

#### Analyze the relationship between incremental revenue and incremental sold unitsfor different promotion types in Hyderabad. Which promotion type led to the highest incremental sold units, and which one generated the highest incremental revenue? What insights can you draw from the balance between the two metrics for this city?

```python
df_hyderabad = df_stores_events[df_stores_events['city'] == 'Hyderabad']
df_hyderabad.head()
```

```python
quantity_bef = df_hyderabad.groupby('promo_type')['quantity_sold(before_promo)'].sum()
quantity_aft = df_hyderabad.groupby('promo_type')['quantity_sold(after_promo)'].sum()

quantity_diff = quantity_aft - quantity_bef
ISU = quantity_diff * 100 /quantity_bef
ISU
```

Insights from Incremental Sold Units:
BOGOF is the Most Effective:

The BOGOF promotion led to the largest increase in sold units, indicating strong customer preference for promotions that offer immediate value.
Expanding BOGOF to more products or extending its duration could further capitalize on its effectiveness.
500 Cashback Shows Strong Appeal:

Cashback offers also performed well, indicating that customers value direct monetary benefits, especially for higher-ticket items.
Further analysis could explore whether cashback promotions were particularly effective for certain customer segments or product categories.
Discount-Based Promotions:

33% OFF performed moderately well, but 50% OFF underperformed, suggesting diminishing returns for deeper discounts.
25% OFF was ineffective and even resulted in a decline in sold units. This could be due to a lack of perceived value or competition from more attractive promotions.
Optimize Promotion Types:

Focus on promotion types that deliver the most value to customers (e.g., BOGOF and Cashback).
Reevaluate or discontinue less effective promotions like 25% OFF.

```python
df_hyderabad['revenue_bef_promo'] = df_hyderabad['base_price(before_promo)'] * df_hyderabad['quantity_sold(before_promo)']
df_hyderabad['revenue_aft_promo'] = df_hyderabad['base_price(after_promo)'] * df_hyderabad['quantity_sold(after_promo)']
df_hyderabad.head(3)
```

```python
rev_bef = df_hyderabad.groupby('promo_type')['revenue_bef_promo'].sum()
rev_aft = df_hyderabad.groupby('promo_type')['revenue_aft_promo'].sum()

IR = (rev_aft - rev_bef) * 100 / rev_bef
IR
```

Insights from Incremental Revenue:
500 Cashback is the Most Effective:

500 Cashback generated the highest incremental revenue, showing that monetary benefits resonate well with customers.
Expanding cashback offers, especially on high-margin or high-demand products, could maximize profitability.
BOGOF Performs Well:

While it didn’t surpass 500 Cashback, BOGOF promotions still generated substantial revenue gains, making it a highly effective strategy.
The success of BOGOF likely depends on product type and customer perception of value.
Discount-Based Promotions Are Ineffective:

All discount-based promotions (25% OFF, 33% OFF, 50% OFF) resulted in revenue losses, with deeper discounts causing smaller losses.
This suggests that straightforward price reductions may not align with customer expectations or profit margins.

```python
data = {
    'promo_type': ISU.index,
    'incremental_sold_units': ISU.values,
    'incremental_revenue': IR.values
}

df_promo = pd.DataFrame(data)

# Preview the DataFrame
print(df_promo)
```

```python
# Create the scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=df_promo,
    x='incremental_sold_units',
    y='incremental_revenue',
    hue='promo_type',
    style='promo_type',
    s=200  # Marker size
)

# Add labels and title
plt.title('Incremental Revenue vs Incremental Sold Units by Promotion Type (Hyderabad)')
plt.xlabel('Incremental Sold Units')
plt.ylabel('Incremental Revenue')

# Display the plot
plt.tight_layout()
plt.show()
```

1. Which promotion type led to the highest incremental sold units?
BOGOF (Buy One Get One Free) led to the highest incremental sold units, as it is the farthest point on the x-axis (incremental sold units: 252).
This indicates that customers were highly responsive to this type of promotion, which likely encouraged bulk purchases due to its perceived value.

2. Which promotion type generated the highest incremental revenue?
500 Cashback generated the highest incremental revenue, as it is the highest point on the y-axis (incremental revenue: 123).
This indicates that cashback promotions were highly effective in increasing total revenue, likely due to their appeal for high-value purchases.

3. Relationship Between Incremental Revenue and Incremental Sold Units
BOGOF (Buy One Get One Free):

Highest incremental sold units, but its incremental revenue is significantly lower than that of 500 Cashback.
This suggests that while BOGOF drives large sales volumes, it might involve lower per-unit profit margins, affecting revenue.
500 Cashback:

High incremental revenue and strong incremental sold units, indicating that cashback promotions provide a good balance between volume and profitability.
This promotion appeals to both customers purchasing larger quantities and those seeking monetary benefits.
Discount-Based Promotions (25%, 33%, and 50% OFF):

All three discount-based promotions resulted in negative incremental revenue, despite some increase in sold units for 33% OFF and 50% OFF.
This suggests that discounts eroded margins and were not effective in driving profitability. 25% OFF even showed negative incremental sold units, indicating poor customer response.

4. Insights from the Balance Between the Two Metrics
BOGOF is Best for Driving Sales Volume:

Promotions like BOGOF are highly effective at driving large quantities sold, making them suitable for fast-moving consumer goods (FMCG) or products with higher inventory levels.
500 Cashback Balances Revenue and Volume:

Cashback promotions are ideal for generating high revenue and decent sales volume, especially for high-value products or price-sensitive customer segments.
Discount-Based Promotions Are Ineffective:

Straight discounts (e.g., 25% OFF, 33% OFF, and 50% OFF) were not effective in either driving sufficient volume or generating revenue, leading to negative profitability.
Tailor Promotions to Goals:

Use BOGOF when the objective is to clear inventory or drive sales volume.
Use 500 Cashback when the goal is to maximize revenue while maintaining healthy sales volume.


#### Analyze the revenue before and after promotions across different product categories in Bengaluru. What trends can be identified in the performance of each category, and how did promotions impact overall revenue in the city?

```python
df_bengaluru = df_stores_events[df_stores_events['city'] == 'Bengaluru']
df_bengaluru.head(3)
```

```python
df_bengaluru['revenue_bef_promo'] = df_bengaluru['base_price(before_promo)'] * df_bengaluru['quantity_sold(before_promo)']
df_bengaluru['revenue_aft_promo'] = df_bengaluru['base_price(after_promo)'] * df_bengaluru['quantity_sold(after_promo)']
df_bengaluru.head(3)
```

```python
df_bengaluru = pd.merge(df_bengaluru, df_dim_products, on='product_code')
df_bengaluru.head(2)
```

```python
cat_rev_bef = df_bengaluru.groupby('category')['revenue_bef_promo'].sum()
cat_rev_aft = df_bengaluru.groupby('category')['revenue_aft_promo'].sum()
print(cat_rev_aft), print(cat_rev_bef)
```

```python
categories = cat_rev_bef.index
bef_promo = cat_rev_bef.values
aft_promo = cat_rev_aft.values

x = np.arange(len(categories))
bar_width = 0.4

plt.figure(figsize=(8, 6))
bars1= plt.bar(x - bar_width/2, bef_promo, width=bar_width, label='Revenue Before Promotion')
bars2 = plt.bar(x + bar_width/2, aft_promo, width=bar_width, label='Revenue After Promotion')

for bar in bars1:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f'{bar.get_height():}', 
             ha='center', va='bottom', fontsize=8, color='black')

for bar in bars2:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f'{bar.get_height():}', 
             ha='center', va='bottom', fontsize=8, color='black')
# Add chart details
plt.title('Revenue Before and After Promotions by Product Category in Bengaluru')
plt.xlabel('Product Category')
plt.ylabel('Revenue')
plt.xticks(x, categories, rotation=45)
plt.legend()

# Display the chart
plt.tight_layout()
plt.show()
```

Insights from the Revenue Analysis:
Combo1:

Revenue increased significantly (from 15.77M to 38.12M), making it the best-performing category.
Promotions on bundles were highly effective in driving customer purchases.
Grocery & Staples:

Revenue showed moderate growth (from 12.26M to 13.86M).
Promotions encouraged slight increases in bulk purchases.
Home Appliances:

Revenue nearly doubled (from 2.18M to 4.12M).
Promotions were effective in pushing high-value purchases.
Home Care:

Revenue increased from 2.10M to 3.18M.
Customers responded well to promotions, likely stocking up.
Personal Care:

Revenue decreased (from 0.57M to 0.39M).
Promotions were ineffective and possibly misaligned with customer preferences.

Overall Impact:
Total revenue increased from 32.89M to 59.67M, with an overall growth of 81.5%.
Promotions were most impactful for Combo1 and Home Appliances categories.
